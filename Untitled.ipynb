```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Land Data

land = pd.read_csv("Land_Temp_data/land_data.csv")
land = land[["Area","Year","Item","Value"]]
land = land[(land["Year"] == 2012) | (land["Year"] == 2022)]
land = land[(land["Item"]=="Country area") | (land["Item"]=="Forest land")]
states1 = land["Area"]
land
```

```python
# Temp Data

temp = pd.read_csv("Land_Temp_data/temp_data.csv")
temp = temp[(temp["Year"] == 2012) | (temp["Year"] == 2022)]
temp = temp.rename(columns={"Country":"Area"})
temp = temp[["Area","Year","Annual Mean"]]
states2 = temp["Area"]
temp
```

```python
# Merged the tables and removed missing values

df = land.merge(temp)
df.drop([370,371,352,353],axis=0,inplace=True)

# df[:20]
```

```python
# Difference in forest coverage from 2012 to 2022

forest = df[(df["Item"] == "Forest land")]
forest1 = forest[forest["Year"]==2012]["Value"]
forest2 = forest[forest["Year"]==2022]["Value"]

total = df[(df["Item"] == "Country area")]
total = total[(total["Year"] == 2012)]["Value"]

diff_land = pd.Series(list((forest2.values - forest1.values)*100/total.values))

diff_land.describe()
```

```python
# Difference in temperature from 2012 to 2022

temp1 = df[(df["Year"] == 2012) & (df["Item"] == "Country area")]["Annual Mean"]
temp2 = df[(df["Year"] == 2022) & (df["Item"] == "Country area")]["Annual Mean"]

diff_temp = pd.Series(list(temp2.values - temp1.values))

diff_temp.describe()
```

```python
# Dealing with Outliers

diff = pd.DataFrame({"Land":diff_land,"Temp":diff_temp})
diff = diff.drop(diff.index[(diff["Land"] >= 2.5) | (diff["Land"] <= -5)])
diff = diff.drop(diff.index[(diff["Temp"] >= 1.2) | (diff["Temp"] <= -0.5)])
diff
```

```python
l_div = pd.Series(list(diff["Land"] - diff["Land"].mean()))
t_div = pd.Series(list(diff["Temp"] - diff["Temp"].mean()))
cov = pd.Series(list(l_div.values * t_div.values))
cov = cov.sum()
cov /= len(diff["Land"])
corr = cov / (diff["Land"].std() * diff["Temp"].std())
corr
```

```python
plt.scatter(diff["Land"],diff["Temp"])
plt.show()
```
